Run the cell below to create or overwrite `app.py` with the complete dashboard code. This file contains all the logic for your Streamlit application.

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# ------------------------------
# Page Configuration
# ------------------------------
st.set_page_config(page_title="Sales & Revenue Dashboard", layout="wide")

st.title("📊 Sales and Revenue Dashboard")
st.markdown("Analyze sales performance, revenue trends, and top products")

# ------------------------------
# File Upload
# ------------------------------
uploaded_file = st.file_uploader(
    "Upload Excel or CSV File",
    type=["csv", "xlsx"]
)

# ------------------------------
# Load Data
# ------------------------------
if uploaded_file is not None:

    try:
        # Read CSV or Excel
        if uploaded_file.name.endswith('.csv'):
            df = pd.read_csv(uploaded_file)
        else:
            df = pd.read_excel(uploaded_file)

        st.success("File uploaded successfully!")

        # ------------------------------
        # Required Columns Check
        # ------------------------------
        required_columns = ['Date', 'Product', 'Category', 'Quantity', 'Price']

        missing_columns = [col for col in required_columns if col not in df.columns]

        if missing_columns:
            st.error(f"Missing columns: {missing_columns}")
            st.stop()

        # ------------------------------
        # Data Processing
        # ------------------------------
        df['Date'] = pd.to_datetime(df['Date'])
        df['Revenue'] = df['Quantity'] * df['Price']

        # ------------------------------
        # Sidebar Filters
        # ------------------------------
        st.sidebar.header("Filters")

        # Date Range Filter
        min_date_available = df['Date'].min().date()
        max_date_available = df['Date'].max().date()

        date_range = st.sidebar.date_input(
            "Select Date Range",
            value=(min_date_available, max_date_available),
            min_value=min_date_available,
            max_value=max_date_available
        )

        if len(date_range) == 2:
            start_date_selected = pd.to_datetime(date_range[0])
            end_date_selected = pd.to_datetime(date_range[1])
        else:
            # Default to full range if only one date is selected or none
            start_date_selected = pd.to_datetime(min_date_available)
            end_date_selected = pd.to_datetime(max_date_available)

        category_filter = st.sidebar.multiselect(
            "Select Category",
            options=df['Category'].unique(),
            default=df['Category'].unique()
        )

        product_filter = st.sidebar.multiselect(
            "Select Product",
            options=df['Product'].unique(),
            default=df['Product'].unique()
        )

        filtered_df = df[
            (df['Date'] >= start_date_selected) &
            (df['Date'] <= end_date_selected) &
            (df['Category'].isin(category_filter)) &
            (df['Product'].isin(product_filter))
        ]

        # ------------------------------
        # KPI Metrics
        # ------------------------------
        total_revenue = filtered_df['Revenue'].sum()
        total_sales = filtered_df['Quantity'].sum()
        avg_order_value = filtered_df['Revenue'].mean()

        col1, col2, col3 = st.columns(3)

        col1.metric("Total Revenue", f"₹{total_revenue:,.2f}")
        col2.metric("Total Sales", f"{total_sales:,}")
        col3.metric("Average Order Value", f"₹{avg_order_value:,.2f}")

        st.markdown("---")

        # ------------------------------
        # Revenue Trend Chart
        # ------------------------------
        revenue_trend = filtered_df.groupby('Date')['Revenue'].sum().reset_index()

        fig_line = px.line(
            revenue_trend,
            x='Date',
            y='Revenue',
            title='Revenue Trend Over Time',
            markers=True
        )

        st.plotly_chart(fig_line, use_container_width=True)

        # ------------------------------
        # Top Products Chart
        # ------------------------------
        top_products = (
            filtered_df.groupby('Product')['Revenue']
            .sum()
            .sort_values(ascending=False)
            .head(10)
            .reset_index()
        )

        fig_bar = px.bar(
            top_products,
            x='Product',
            y='Revenue',
            title='Top 10 Products by Revenue',
            text_auto=True
        )

        st.plotly_chart(fig_bar, use_container_width=True)

        # ------------------------------
        # Category-wise Revenue
        # ------------------------------
        category_revenue = (
            filtered_df.groupby('Category')['Revenue']
            .sum()
            .reset_index()
        )

        fig_pie = px.pie(
            category_revenue,
            names='Category',
            values='Revenue',
            title='Revenue by Category'
        )

        st.plotly_chart(fig_pie, use_container_width=True)

        # ------------------------------
        # Display Data Table
        # ------------------------------
        st.subheader("Sales Data")
        st.dataframe(filtered_df)

    except Exception as e:
        st.error(f"Error: {e}")

else:
    st.info("Please upload a CSV or Excel file to start.")

    st.subheader("Sample Dataset Format")

    sample_data = pd.DataFrame({
        'Date': ['2026-01-01', '2026-01-02'],
        'Product': ['Laptop', 'Mobile'],
        'Category': ['Electronics', 'Electronics'],
        'Quantity': [5, 10],
        'Price': [50000, 20000]
    })

    st.dataframe(sample_data)

# ------------------------------
# Run Instructions
# ------------------------------
# Save this file as app.py
# Install required libraries:
# !pip install streamlit pandas plotly openpyxl
#
# Run the dashboard:
# streamlit run app.py
```
